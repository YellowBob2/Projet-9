# Probleme du sac-à-dos:
étant donné un ensemble d’objets ayant chacun une valeur et un
poids, et une capacité B, retourne un sous-ensemble d’objets de poids au plus B et de valeur
total maximale.

```python
from copy import copy
import random
import numpy as np
from matplotlib import pyplot as plt
```

```python
def somme(tab):
    resval=0
    resp=0
    for _,val,poids in tab:
        resval=resval+val
        resp=resp+poids
    return (resval,resp)

def recursac(tab,B,tailletab,i,tabactuel,valmax,tabmax):
    if(i==tailletab):
        val,poids=somme(tabactuel)
        if(poids<=B and val>valmax):
            tabmax=tabactuel
            valmax=val
        return tabmax
    res1=recursac(tab,B,tailletab,i+1,tabactuel,valmax,tabmax)
    tabpro=copy(tabactuel)
    tabpro.append(tab[i])
    res2=recursac(tab,B,tailletab,i+1,tabpro,valmax,tabmax)
    val1,poids1=somme(res1)
    val2,poids2=somme(res2)
    if(poids1<=B and val1>valmax):
        tabmax=res1
        valmax=val1
    if(poids2<=B and val2>valmax):
        tabmax=res2
        valmax=val2
    return tabmax
#Tuple[int,float,float]
#(idobjet,prix,poids)
#B est le poids max du sac
#prend en parametre un tableau de Tuple[int,float,float] correspondant à (idobjet,prix,poids) et le poids max du sac
def sac(tab,B):
    return recursac(tab,B,len(tab),0,[],0,[])
```

```python
t=[(1,5,3),(2,3,4),(3,3,9)]
print(sac(t,12))
```

```python
def randomtab(taille,minval,maxval,minpoids,maxpoids):
    tab=[]
    for i in range(0,taille):
        tab.append((i,round(minval+random.random()*(maxval-minval),2),round(minpoids+random.random()*(maxpoids-minpoids),2)))
    return tab
def test(taille,poidsmaxtotal,minval,maxval,minpoids,maxpoids):
    tab=randomtab(taille,minval,maxval,minpoids,maxpoids)
    print("----------------------------")
    print("LES OBJETS (idobjet,prix,poids)")
    print(tab)
    res=sac(tab,poidsmaxtotal)
    print("----------------------------")
    print("VOICI CE QU'IL FAUT METTRE DANS LE SAC")
    print(res)
    return
def testdetaille(taille,poidsmaxtotal,minval,maxval,minpoids,maxpoids):
    tab=[]
    for i in range(1,taille+1):
        elem=(i,round(minval+random.random()*(maxval-minval),2),round(minpoids+random.random()*(maxpoids-minpoids),2))
        print("----------------------------")
        print("ETAPE "+str(i))
        print("L'OBJET AJOUTE (idobjet,prix,poids)")
        print(elem)
        tab.append(elem)
        res=sac(tab,poidsmaxtotal)
        print("VOICI CE QU'IL FAUT METTRE DANS LE SAC")
        print(res)
    print("----------------------------")
    print("FIN")
    print("----------------------------")
    return
```

# Budget participatif:
Un algorithme de budget participatif prend en entrée une liste de projets proposés par les
habitants, les coûts de chacun de ces projets, les préférences des votants sur les projets, et le
budget disponible B. Il retourne un sous-ensemble des projets dont la somme des coûts est au
plus B.

```python
def BP(projet,tab,B,tabvotes):
    """tab est une liste avec des tuples de la forme (idprojet,nbvotes,prix) mettre nbvotes à 0 au debut
    tabvotes est une liste d'entier avec des id de projet.
    """
    array=[]
    array.append(np.zeros(len(projet)))
    for i in range(0,len(tabvotes)):
        x,y,z=tab[tabvotes[i]]
        tab[tabvotes[i]]=(x,y+1,z)
        for j in range(0,len(projet)):
            array.append(np.zeros(len(projet)))
            array[i+1][j]=array[i][j]
            if(j==tabvotes[i]):
                array[i+1][j]=array[i+1][j]+1
    plt.figure(figsize=(12.8,9.6))
    plt.title("")
    plt.xlabel("Temps") 
    plt.ylabel("Nombre de votes")
    x = np.linspace(0, len(tabvotes)-1,len(tabvotes))
    for k in range(0,len(projet)):
        array2=[]
        for l in range(0,len(tabvotes)):
            array2.append(array[l][k])
        plt.plot(x,array2,label=projet[k])
    plt.xlim(0,len(tabvotes))
    plt.ylim(ymin=0)
    plt.legend(loc='best')
    plt.show()
    res=recurBP(tab,B,len(tab),0,[],0,[])
    for a in res:
        b,_,_=a
        print(projet[b])
    return

def recurBP(tab,B,tailletab,i,tabactuel,votesmax,tabmax):
    if(i==tailletab):
        votes,prix=somme(tabactuel)
        if(prix<=B and votes>votesmax):
            tabmax=tabactuel
            votesmax=votes
        return tabmax
    res1=recurBP(tab,B,tailletab,i+1,tabactuel,votesmax,tabmax)
    tabpro=copy(tabactuel)
    tabpro.append(tab[i])
    res2=recurBP(tab,B,tailletab,i+1,tabpro,votesmax,tabmax)
    votes1,prix1=somme(res1)
    votes2,prix2=somme(res2)
    if(prix1<=B and votes1>votesmax):
        tabmax=res1
        votesmax=votes1
    if(prix2<=B and votes2>votesmax):
        tabmax=res2
        votesmax=votes2
    return tabmax
```

```python
projet=["Parc","Piscine Municipale","Zoo"]
tab=[(0,0,40000),(1,0,50000),(2,0,100000)] #(idprojet,nbvotes,prix) mettre nbvotes à 0 au debut
tabvotes=[0,0,1,2,1,2,1,0,0]
B=100000
BP(projet,tab,B,tabvotes)
```
